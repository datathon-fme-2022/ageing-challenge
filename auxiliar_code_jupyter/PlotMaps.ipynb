```python
import plotly.express as px
import pandas as pd



us_cities = pd.read_csv(
    "https://raw.githubusercontent.com/plotly/datasets/master/us-cities-top-1k.csv"
)

us_cities.head()

```

```python
dicctionary = {"Name":[], 
               "Recorded Text":[],
               "lat":[],
               "lon": []
               }

old_peop = [
    "Rogelia", "Help, I am dying", 41.38879, 2.15899
]

df = pd.DataFrame(columns=["Name", "Recorded Text", "lat", "lon"])
df = df.append({"Name":"Rogelia", "Recorded Text":"Help, I am dying", "lat":41.38879, "lon":2.15899}, ignore_index=True)
df
```

```python

fig = px.scatter_mapbox(
    df,
    lat="lat",
    lon="lon",
    hover_name="Name",
    hover_data=["Recorded Text"],
    color_discrete_sequence=["red"],
    zoom=15,
    height=500,
)
fig.update_traces(marker={'size': 15})
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.show()
```
